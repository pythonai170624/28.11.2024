```python
import numpy as np 
import pandas as pd
from IPython.display import display

print('========== read mpg miles per gallon data on cars =======================')
mpg_df = pd.read_csv(r"mpg.csv")

#display(mpg_df)
#display(mpg_df.describe())
#display(mpg_df.info())

mpg_df.groupby('model_year')['weight'].mean()
mpg_df.groupby('model_year')[['weight', 'mpg']].mean()
#mpg_df.groupby('model_year')[['weight', 'mpg']].mean().reset_index()
mpg_df.groupby('model_year')[['weight', 'mpg']].mean().rename(columns={'weight': 'mean_weight', 'mpg':'mean_mpg'})
mpg_df.groupby('model_year')[['weight', 'mpg']].mean().rename(columns={'weight': 'mean_weight', 'mpg':'mean_mpg'})
mpg_df.groupby('model_year')['weight'].mean().sort_values()

# mpg_df.groupby('model_year')['horsepower'].mean().sort_values() # error
mpg_df['horsepower'] = pd.to_numeric(mpg_df['horsepower'], errors='coerce')
#mpg_df.info()

########## The groupby operation is not just about creating a multi-index. Its primary role is aggregation,
########       transformation, or filtering of data based on groups. It automatically groups the data and 
#######         applies operations like sum, mean, count, etc.
########## Explicitly creating a multi-index (e.g., with set_index or pd.MultiIndex) is for structuring the data into
#########    a hierarchical format without performing any grouping or aggregation.


mpg_df.groupby('model_year')[['horsepower', 'weight']].mean().sort_values(by='weight')

mpg_df.groupby('model_year')[['horsepower']].value_counts()  # how many ikn group + same horsepower, i.e. 70 8
mpg_df.groupby('model_year')['model_year'].value_counts()  # how many in each group?

mpg_df.groupby('model_year')['horsepower'].nlargest(5)

mpg_df.groupby('model_year')[['horsepower']].value_counts()

mpg_df.groupby('model_year').max()
mpg_df.groupby('model_year').min()
#mpg_df.groupby('model_year').mean()

# dropping name column and then mean -- all columns must be numeric for mean
display(mpg_df.groupby('model_year')[mpg_df.columns.difference(['name'])].mean())

print(mpg_df.groupby('model_year').max()['mpg'])

print(mpg_df.groupby('model_year').max().loc[74])

mpg_df.groupby('model_year')['horsepower'].nlargest(5)
```

```python
beer = pd.read_csv(r"RetailSales_BeerWineLiquor.csv")
#display(beer)
display(beer.info())
beer['DATE'] = pd.to_datetime(beer['DATE'])
display(beer.info())
beer.set_index('DATE', inplace=True)
display(beer.resample(rule='A').sum().head(5))  # resample = group by the same year
display(beer.resample(rule='A').mean().head(5))
# resample: Groups the data into specified time intervals based on the rule provided.
# rule='A':
# 'A' stands for "Annual", so the data will be resampled into yearly intervals.
# The default behavior aggregates data at the end of each year (December 31st).
# mean(): After grouping the data into yearly intervals, this computes the mean for each group.

#'A'	Year-end frequency	Dec 31, 2020; Dec 31, 2021
#'AS'	Year-start frequency	Jan 1, 2020; Jan 1, 2021
#'Q'	Quarter-end frequency	Mar 31, 2020; Jun 30, 2020
#'QS'	Quarter-start frequency	Jan 1, 2020; Apr 1, 2020
#'M'	Month-end frequency	Jan 31, 2020; Feb 29, 2020
#'MS'	Month-start frequency	Jan 1, 2020; Feb 1, 2020
#'W'	Week-end frequency	Jan 5, 2020; Jan 12, 2020
#'W-SUN'	Week ending on Sunday (default)	Jan 5, 2020; Jan 12, 2020
#'W-MON'	Week ending on Monday	Jan 6, 2020; Jan 13, 2020
#'D'	Daily frequency	Jan 1, 2020; Jan 2, 2020
#'H'	Hourly frequency	01:00, 02:00, 03:00
#'T' or 'min'	Minutely frequency	01:01, 01:02, 01:03
#'S'	Secondly frequency	01:01:01, 01:01:02
#'L'	Millisecond frequency	01:01:01.001, 01:01:01.002
#'U'	Microsecond frequency	01:01:01.000001, 01:01:01.000002
#'N'	Nanosecond frequency	01:01:01.000000001, 01:01:01.000000002

beer = pd.read_csv(r"RetailSales_BeerWineLiquor.csv")
beer['DATE'] = pd.to_datetime(beer['DATE'])
display(beer.info())
display(beer[beer['DATE'] > '2000-01-01'])
beer = beer[beer['DATE'] > '2000-01-01']
beer.set_index('DATE', inplace=True)
display(beer.resample(rule='A').sum().head(5))  # resample = group by the same yearנ
```
