```python
print('========== read mpg miles per gallon data on cars =======================')
mpg_df = pd.read_csv(r"mpg.csv")

mpg_mean = mpg_df['mpg'].mean()
weight_mean = mpg_df['weight'].mean()
mpg_max = mpg_df['mpg'].max()
weight_std = mpg_df['weight'].std()

print(mpg_mean)
print(weight_mean)
print(mpg_max)
print(weight_std)

columns = ['stats', 'mpg', 'weight']
data = [
    ['max', mpg_max, None],
    ['mean', mpg_mean, weight_mean],
    ['std', None, weight_std]
]

stats_df = pd.DataFrame(data=data, columns=columns)
display(stats_df)

#display(mpg_df.agg({'mpg': ['max', 'mean', 'min'], 'weight': ['std', 'mean']}).transpose())

# before-- 
# 1 agg function
# mpg_df.groupby(['cylinders', 'model_year']).max(numeric_only=True)
# multiple agg functions on each column ...
mpg_df.groupby(['cylinders', 'model_year']).agg({'weight': ['max', 'min'], 'mpg': 'mean', 'horsepower': 'min'})
```
